 #### 1의 보수 
 ex) 8bit
     +10: 00001010  
     -10: 11110101
      
     3 - 1 -> (+3) - (+1)  
           -> (+3) + (-1)   
           
           +10 : 00001010  
           -10 : 11110101  
           서로 비트를 반전시켜 음수의 개념을 만듬  
           즉 크기가 같고 부호가 다른 두 수를 더해서 11111111을 만듬
           
           의문점 11111111이 어떻게 0이됨? 
           맨 앞자리가 1이므로 컴퓨터는 위의 숫자를 음수로 인식함.
           1의 보수를 사용하는 컴퓨터는 이를 다시 재 보수화 시킴
           즉 11111111은 전부 0으로 바뀜.
           그래서 값이 0이 나옴.
           
             
           +3 : 00000011  
           +1 : 00000001  
           -1 : 11111110  
           
           +3 + (-1) = 0000011  
                    +ㅣ1111110  
                    = 10000001  
                    여기서 8비트의 값을 넘어가는 1000000 값이 발생함  
                    이를 캐리아웃값 이라고함.  
                    이후 반드시 결과값에 캐리아웃값 1을 다시 더해줌.    
                    
  #### 2의 보수
  ex) 8bit
     +10: 00001010  
     -10: 11110110
     1의 음수 개념과는 다름.
     크기가 같고 부호가 다른 두 수를 더해서 00000000을 만듬
     
      +3 + (-1) = 0000011  
               +ㅣ1111111   
               = 00000010 (1의 보수방식과는 다르게 캐리어를 무시)
               
               
  ### Tip!
  보통의 컴퓨터는 2의 보수 방식을 사용한다.
  이때 어떤 양의 정수값의 부호를 쉽게 바꾸는 방법.
  
  10 : 0000 1010 일 때  
  오른쪽 부터 시작하여 왼쪽으로 이동하며   
  1을 찾음. 1을 찾았다면 멈추고 1의 왼편은 전부 비트를 반전시키고 1의 오른편은 그대로 쓴다.  
  
  10 : 0000 1010  
              ^  
              1의 위치  
              
 -10 : 1111 0110  
 
 
 #### bit 크기
