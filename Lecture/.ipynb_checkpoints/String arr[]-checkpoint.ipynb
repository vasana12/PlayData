# String arr[]  
---
  
**객체로 구성된 배열의 메모리 구조**  
- 사용법  

``` 
String arr1 [] = new String [100];  

```  
    
**데이터 구조**  
```
String arr1[] = {"a", "b", "c"};

```
일 때, 데이터의 구조를 살펴보자.  

  
배열 자체는 참조형식의 데이터인 것은 이해할 것이다.  
이때, 배열에 참조형식의 데이터를 집어 넣으면 어떻게 될까?  
    
- int 배열을 먼저 보면  
```
int arr2 [] = {1,2,3};
```
  
일 때, arr2 자체는 배열의 주소를 가르키게 된다. (1차 참조)  
  
  arr2 : ex) (주소)100 저장  
            [100] : 1, [104] : 2, [108] : 3 해당 주소에 가면 실제 값이 저장되 있다.  
       

### 하지만 객체를 저장하는 배열일 경우는 다음과 같다.  
  
  
---
 arr1 : ex) (주소)100 저장  
            [100] : [200], [104] : [210], [108] : [220]  
            [200] : "a", [210] : "b", [220] : "c"    
            가 저장 되있다. 이는 2중 참조 형식이라 생각하면 된다.  
  
  
  
  
 - 이 점에 의해서 객체를 단순 논리 연산자로 비교할 때 문제가 발생한다.  
  가령 String a = "a"; 라 선언 했을 때,  
  a == "a";   
  하면 값은 당연히 true일 것이라 생각했다면, **틀렸다.**     
  값은 false이다.    
    
  
  
   - 이유는 a는 객체형식이기 때문에 ==를 사용하면 a의 주소를 반환하여 "a"의 주소와 비교하게 된다.   
    즉 String a의 주소 == "a"의 주소 를 비교하기 때문에 false를 반환한다.  
    
  이 때는 a.equals("a");와 같이 String 클래스의 equals()메소드를 이용하여 비교한다.
