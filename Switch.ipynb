# Switch  
---

자바에서 스위치 문의 매개변수는 숫자형과 문자 상수만 사용한다. (1.8버전? 이후로는 문자열 상수도 가능)  
주의할 점은 스위치는 값에 의한 분기문이다. 
###### if문 처럼 조건으로 분기하지 못한다.(사전 처리 필요)  
```
int a = 0;

switch (a) 
{
  case 1:
    실행문;
    break;
  case 2:
    실행문;
    break;
  default:
     실행문;
    //break; 생략 가능
}
```

#### Tips!   
- Tip1  

```
char a = 'a'; //이거나  
char a = 'A'; //일때 같은 명령문을 수행하고 싶다면,  

switch (a) 
{
  case 'a':
  case 'A':
    실행문;
    break;
  default:
     실행문;
    //break; 생략 가능
}
와 같이 작성이 가능하다.
```
  
- Tip2     

숫자 범위로 switch 문을 분기하고 싶을 때!  

```

int a = 50;
char b;

if(a < 30){
    b = '미';
}else if(a <40){
    b = '양';
}
    ....
```
처럼 이프문을 이용해서 할 필요 없이 다음과 같은 작업을 수행하면 수월하다.  

#### 형변환을 이용하는 것!

우선 int c를 선언하자  
이후 c에 다음과 같은 결과를 대입한다.  

```
c  = (int)(a / 10);
```
이렇게 하면 90점대는 9로 범위를 나타내고   
80점대는 8로 범위를 나타낼수 있다.  

```
int a =50;
int c = (int)(a / 10);
char b

switch (c)
{
    case 10:
    case 9:
    b = '수';
    break;
    case 8:
    b = '우';
    break;
    case 7:
    b = '양';
    break;
}
```

### 이런 팁으로 이프문없이 스위치문을 사용할 수 있다. 잔머리 필수!
