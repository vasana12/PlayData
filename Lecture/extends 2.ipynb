### 메소드 다형성  

```
abstract class Sender ( String name, String title ){
    String name, title;
    
    Sender(String name, String title){
        this.name = name;
        this.title = title;
    }
    abstract messageSender(String contens, String recipient); // 추상 메소드 오버라이딩
}
```
  
  위와 같은 추상 클래스가 있다. 그리고 위 추상 클래스를  
  상속받아 구현하는 자식클래스 2개가 다음과 같다.
  
```  
class Phonesender (String name, String title, String phoneNum, String recipient) extends  Sender {
    String phoneNum, recipient;
    
    Phonesender(String name, String title){
        this.name = name;
        this.title = title;
    }
    messageSender(String contens, String recipient); // 추상 메소드 오버라이딩
}

```

main 클래스

```
public class Main {

    //메소드 
    void message(Sender obj){  // 메소드의 매개변수를 주목하라. 부모 클래스의 타입으로 받는다.
        System.out.print(obj.name);
        System.out.print(obj.title);
        Sender.messageSender(String contens, String recieptent);
    }
    
    //진입점
	public static void main(String[] args) {
		Phonesender obj1 = new Phonesender("김경우", "밥먹자", "010-202-2222", "헝길");
        EmaileSender obj2 = new EmaileSender("김경우", "밥먹자", "010_2202-2030", "길헝"");
        //두개의 자식 클래스 객체 생성
        
	
        message(obj1);
        message(obj2);

	}
}
```
메세지 메소드 호출  
매개변수는 각각의 자식클래스를 받는다.  
        이는 형식 매개변수를 부모클래스로 했기 때문에 가능하다.  
        이는 다형성의 의미를 갖는다.
          
          
**주의할 점**  
 형식 매개변수를 부모 클래스로 선언하면    
자식클래스 객체를 해당 매개변수 인자로 전달이 가능하다.  
하지만 반대의 상황은 (형식 매개변수 : 자식 , 인자 : 부모) 불가능하다.  
  
또 message()메소드 안에서 접근 가능한 인자들은  
부모 클래스에 정의된 인자와 메소드이다.  
 즉 자식클래스의 객체를 인자로 넘겨 주어도  
해당 메소드에서 사용이 가능한 인자는  
부모 클래스에서 정의된 name,title, messageSender뿐이다.
