# Shift Operators  
---
  
  - 사용법  
    
  정수 << 비트수 or 정수 >>비트수  
  (정수 <<< 비트수) or 정수 >>>비트수  
  
    
   
   
### 좌측 shift(산술 연산) 
**shift 산술 연산자는 피연산자의 부호(맨 왼쪽의 bit)을 제외하고 연산한다. **
-  ex)  
   +10 : 0000 1010  
   -10 : 1111 0110  
   
   위 두 비트를 좌측으로 2비트 시프트 시키면 다음과 같이 나온다.  
   +10 -> [0]010 10(00) (+40) (좌측으로 2비트 옮기고 오른쪽 빈공간 2개를 0으로 채움)  
   -10 -> [1]101 10(00) (-40) (좌측으로 2비트 옮기고 오른쪽 빈공간 2개를 0으로 채움)  
   
     
### 우측 shift(산술 연산)
**shift 산술 연산자는 피연산자의 부호(맨 왼쪽의 bit)을 제외하고 연산한다. **
- ex)
    +10 : [0]000 1010
    -10 : [1]111 0110
    
   위 두 비트를 우측으로 2비트 시프트 시키면 다음과 같이 나온다.  
   +10 -> [0] (00)0 0010 (40) (우측으로 2비트 옮기고 왼쪽 빈공간 2개를 **부호 비트**로 채움)  
   -10 -> [1] (11)1 1101 (-40) (우측으로 2비트 옮기고 왼쪽 빈공간 2개를 **부호 비트**로 채움)  
   
  **산술 연산은 말 그래로 산술을 하기 위한 연산이다.**
  
    
       
  
### 좌측 shift(논리 연산)
**shift 산술 연산자는 피연산자의 부호(맨 왼쪽의 bit)도 같이 연산한다. ** 
- ex)
    +10 : 0000 1010
    -10 : 1111 0110
    
     위 두 비트를 좌측으로 2비트 시프트 시키면 다음과 같이 나온다.  
   +10 -> 0010 10(00) (40) (좌측으로 2비트 옮기고 오른쪽 빈공간 2개를 0으로 채움)  
   -10 -> 1101 10(00) (-40) (좌측으로 2비트 옮기고 오른쪽 빈공간 2개를 0으로 채움)  
   
   
 
### 우측 shift(논리 연산)
**shift 산술 연산자는 피연산자의 부호(맨 왼쪽의 bit)도 같이 연산한다. ** 
- ex)
    +10 : 0000 1010
    -10 : 1111 0110
    
   위 두 비트를 우측으로 2비트 시프트 시키면 다음과 같이 나온다.  
   +10 -> (00)00 0010 (40) (우측으로 2비트 옮기고 왼쪽 빈공간 2개를 **0**으로 채움)  
   -10 -> (00)11 1101 (-40) (우측으로 2비트 옮기고 왼쪽 빈공간 2개를 **0**으로 채움)  
   
#### 논리연산을 왜 사용하는가?  
- 논리 연산은 cpu 메모리인 레지스터에서 데이터를 옮기기 위해 사용한다.  
 이 때 좌측 shift논리 연산과 좌측 shift 산술연산은 결과가 같다.!  
 따라서 java에서는 좌측 논리연산은 따로 만들지 않는다.  

### 표기법  
  
- **산술 연산**   
  정수 << 비트수 : 좌측 shift (산술)  
  정수 >> 비트수 : 우측 shift (산술)  
    
      
- **논리 연산**  
  정수 >>> 비트수 : 우측 shift (논리)
