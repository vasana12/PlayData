# for loop
---

**Loop의 구조  **  
- 초기식  
 조건식  
 반복처리할 내용  
 증감식  
   
  
- **1.형식**
      for(초기식 ; 조건식; 증감식)
      {
          반복할 문장;
      }  
      
- **2.형식**
       for( ; ; )  
       {  
           반복할 문장;  
       }    
       
       이 형식은 while (true) 와 같은 형식으로   
       for문 안에 if문과 break를 이용하여 무한루프를 탈출토록 한다.  
       
- **3.형식 : 배열을 이용 ** 
   
         ```
         데이터타입 배열명[] = {배열 원소들};
         
         for( 배열의데이터타입 변수명 : 배열이름)
         {
             반복 문장
         }
         
         ex)
         int arr[] = {1,2,3,4,5};
         
         for( int num : arr)
         {
             print(num);
         }
         //12345
         ```
- **다중 반복문 일 때 한번에 탈출 하는 방법**  
    라벨을 이용하자  
      
    사용법  
    ```
    식별자이름 : //사용자가 라벨 이름을 설정
         for( ~~)
             {
             for(~~)
                 {
                     if(~~)
                     {
                         break 식별자이름;
                     }
                 }
             }
             
      ex)  
      outerLoop:
		for(int row = 0; row <3; row ++) {
			for(int col = 0; col <5; col ++) {
				if(row == 1 && col ==4) {
					break outerLoop;
				}
				out.println("(" + row + "," + col + ")");
			}
		}
    ```
