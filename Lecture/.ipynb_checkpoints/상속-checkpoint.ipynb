# 상속  
---  
 
**클래스 상속**  

 ```
 class 부모클래스명{
   
     부모클래스명(){      // 부모 클래스 생성자
 
     }
   ...
 }
 class 자식클래스명 extends 부모클래스명 { 
   
     자식클래스명(){      // 자식클래스 생성자
   
         //super() 부모 클래스 생성자 호출문
     }
   ...
}
```
  
자바는 단일 상속을 원칙으로 함.  
  
클래스간의 상속 시 조심해야 할 문제가 있다.  
바로 생성자 개념이다.  

자식클래스는 상속받은 부모 클래스의 필드나 메소드에 (접근제한자에 따라)접근이 가능하다.  
이는 생성자도 가능하기 때문에 두 클래스의 객체를 생성했을 때, 주의해야한다.  
  
**중요한 점**  
자식 클래스의 생성자가 default이든 생성자를 명시적으로 선언했든,
생성자 구문에서 맨처음 실행되는 것은    
**super();** 구문이다.  
  
### **이 메소드는 부모 클래스의 생성자를 불러오는 메소드이다.**    
즉 부모 클래스의 객체를 만들지 않고,  
자식클래스의 객체를 선언하면, 두 클래스의 생성자가 모두 실행된다.  
  
실행 순서는 부모 클래스 생성자 -> 자식 클래스 생성자 순으로   
각 클래스의 필드들이 초기화 된다.  
  
이 때 문제가 발생하지 않게 하기위한 일종의 양식이 있다.  
  
### 부모클래스 선언 양식
ex)
```
class parents {
    int a, b;  
      
    parents(){                          //default 생성자를 명시하는게 좋다
    }
    parents( int a, int b){             //부모 필드 초기화
        this.a = a;
        this.b = b;
    }

}

class child{
    int c, d;
    child(int a, int b, int c, int d){  // 부모 클래스의 필드가 생성자를 사용하여 필드를 초기화 했을 때!  
        super(a, b);                    // 명시적으로 (spuer()를 이용하여)  
        this.c = c;                     // 부모 클래스의 생성자에 해당 필드 값을 주어 초기화 시켜 주어야 한다.  
        this.d = d;
    }
}
```

# 수식 키워드
---
### final 키워드
1. final 필드    
    필드를 final로 수식하면 해당 필드는 값변경이 선언 이후에 불가능하다.  
    이 필드는 클래스에서 선언 할 때(필드 생성ㅇ), 혹은 생성자 안에서만 초기화가 가능하다.  

2. final 메소드  
    해당 메소드는 오버라이딩이 불가능 하다.  
    
3. final 클래스  
    해당 클래스는 상속이 불가능 하다.  
  
  
### abstract 키워드 (추상)
추상 메소드 혹은 추상 클래스는 실체 (인스턴스, 객체)가 없다.  
해당 메소드, 클래스는 **상속**받을 클래스를 위한 개념이다.  
추상적인 메소드와 클래스는 자체적인 실체는 없지만,  
**해당 클래스를 *상속받은*  자식 클래스에서**   
부모 클래스의 추상 메소드를 직접 구현하여  
오버라이딩하게 된다.  
자식 클래스는 상속받은 추상 클래스의 추상 메소드를   
무조건 구현해야 하니 주의할 것.  
  
인터페이스와 비슷한 개념이라 생각하면 될듯
